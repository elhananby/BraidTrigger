```python
from zipfile import ZipFile
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
braidz_file = '/mnt/Experiments/Calibration/20230626_121910_pixel_to_direction_calibration_.braidz'

with ZipFile(braidz_file) as archive:
    df = pd.read_csv(archive.open("kalman_estimates.csv.gz"), comment="#", compression="gzip")
```

```python
from scipy.stats import circmean
%matplotlib widget
headings = []
fig, axs = plt.subplots(ncols=2, nrows=1)
for idx, grp in df.groupby("obj_id"):
    axs[0].plot(grp.x, grp.y)
    heading = np.arctan2(grp.yvel, grp.xvel)
    axs[1].plot(heading)
    headings.append(circmean(heading, low=-np.pi, high=np.pi))
plt.show()
```

```python
fig = plt.figure()
plt.scatter(list(range(len(headings))), headings)
plt.show()
```
